```python
import matplotlib.pyplot as plt
import pandas as pd

# Load the CSV
file_path = "/content/drive/MyDrive/PDV Lab Test/PDV Lab Test/book_spider/books.csv"
df = pd.read_csv(file_path)

# Clean the price column (remove '£' and convert to float)
df['price'] = df['price'].replace('£', '', regex=True).astype(float)

# --- Visualization 1: Rating distribution ---
rating_counts = df['rating_value'].value_counts().sort_index()
plt.figure(figsize=(7,5))
plt.bar(rating_counts.index, rating_counts.values)
plt.title("Number of Books by Rating")
plt.xlabel("Rating (1 to 5 stars)")
plt.ylabel("Number of Books")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 2: Average price by rating ---
avg_price = df.groupby('rating_value')['price'].mean()
plt.figure(figsize=(7,5))
plt.bar(avg_price.index, avg_price.values)
plt.title("Average Price by Rating")
plt.xlabel("Rating (1 to 5 stars)")
plt.ylabel("Average Price (£)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 3: Availability distribution ---
availability_counts = df['availability'].value_counts()
plt.figure(figsize=(6,4))
plt.barh(availability_counts.index, availability_counts.values)
plt.title("Availability of Books")
plt.xlabel("Count")
plt.ylabel("Availability Status")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mount and read the CSV (adjust path if needed)
file_path = "/content/drive/MyDrive/PDV Lab Test/PDV Lab Test/book_spider/books.csv"
df = pd.read_csv(file_path)

# Display info
print("✅ Dataset Overview")
print(df.info(), "\n")

print("🔹 First 5 Rows:")
print(df.head(), "\n")

print("🔹 Basic Statistics:")
print(df.describe(include='all'), "\n")

# Clean price column
df['price'] = df['price'].replace('£', '', regex=True).astype(float)

# --- Visualization 1: Number of rows per column (non-null count) ---
non_null_counts = df.count()
plt.figure(figsize=(8,5))
plt.bar(non_null_counts.index, non_null_counts.values, color='teal')
plt.title("Non-Null Count per Column")
plt.xticks(rotation=45)
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 2: Missing value count ---
missing_counts = df.isnull().sum()
plt.figure(figsize=(8,5))
plt.bar(missing_counts.index, missing_counts.values, color='salmon')
plt.title("Missing Values per Column")
plt.xticks(rotation=45)
plt.ylabel("Count of Missing Values")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 3: Rating distribution ---
rating_counts = df['rating_value'].value_counts().sort_index()
plt.figure(figsize=(7,5))
plt.bar(rating_counts.index, rating_counts.values, color='skyblue')
plt.title("Number of Books by Rating")
plt.xlabel("Rating (1–5 stars)")
plt.ylabel("Number of Books")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 4: Average price by rating ---
avg_price = df.groupby('rating_value')['price'].mean()
plt.figure(figsize=(7,5))
plt.bar(avg_price.index, avg_price.values, color='orange')
plt.title("Average Price by Rating")
plt.xlabel("Rating (1–5 stars)")
plt.ylabel("Average Price (£)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 5: Availability distribution ---
availability_counts = df['availability'].value_counts()
plt.figure(figsize=(6,4))
plt.barh(availability_counts.index, availability_counts.values, color='green')
plt.title("Availability of Books")
plt.xlabel("Count")
plt.ylabel("Availability Status")
plt.show()

# --- Visualization 6: Price Distribution ---
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=15, color='purple', edgecolor='black', alpha=0.7)
plt.title("Distribution of Book Prices")
plt.xlabel("Price (£)")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Visualization 7: Price vs Rating scatter ---
plt.figure(figsize=(8,5))
plt.scatter(df['rating_value'], df['price'], color='red', alpha=0.6)
plt.title("Price vs Rating")
plt.xlabel("Rating (1–5 stars)")
plt.ylabel("Price (£)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# --- Visualization 8: Top 10 Most Expensive Books ---
top_10_expensive = df.nlargest(10, 'price')[['title', 'price']]
plt.figure(figsize=(10,6))
plt.barh(top_10_expensive['title'], top_10_expensive['price'], color='darkblue')
plt.title("Top 10 Most Expensive Books")
plt.xlabel("Price (£)")
plt.ylabel("Book Title")
plt.gca().invert_yaxis()
plt.show()

# Summary of dataset
print("✅ Total Rows:", df.shape[0])
print("✅ Total Columns:", df.shape[1])
print("\n✅ Column-wise non-null counts:\n", df.count())
print("\n✅ Missing values:\n", df.isnull().sum())
```
